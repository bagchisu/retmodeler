```python

from retmodeler.asset import Asset, AssetClass, InvestmentType
from retmodeler.expense import Expense
from retmodeler.modeler import Model
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
mm = AssetClass("mm1", 0.03)
lc = AssetClass("lc1", 0.07)
it1 = InvestmentType('t1', False, False, False)
a1 = Asset("savings", 500000, it1, mm)
a2 = Asset("stock", 4500000, it1, lc)
a2.set_reinv_asset(a1,.7)
assets = [a1, a2]
e1 = Expense("expenses", 200000, [a1, a2], 0.03)
expenses = [e1]

model = Model(1964, assets, expenses, 2025)
yearly_values = model.simulate(95)
```

```python
sns.lineplot(yearly_values)
```
