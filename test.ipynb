```python
from retmodeler.asset import Asset, AssetClass, INV_BROKERAGE, INV_401K, INV_ROTH
from retmodeler.expense import Expense, CalculatedExpense
from retmodeler.modeler import Model
from retmodeler.tax import TaxRate
import seaborn as sns
import pandas as pd
```

```python
infl = 0.03
mm = AssetClass("mm1", 0.04, True)
lc = AssetClass("lc1", 0.08, False)
a1 = Asset("savings", 500e3, 0, INV_BROKERAGE, mm)
a2 = Asset("stockBrok", 0.9e6, 450e3, INV_BROKERAGE, lc)
a3 = Asset("stock401k", 1.1e6, 550e3, INV_401K, lc)
a2.set_reinv_asset(a1,.7)
a3.set_reinv_asset(a1,.7)
assets = [a1, a2, a3]
e1 = Expense("expenses", 100e3, assets, infl)
expenses = [e1]
tax_rate = TaxRate(infl)
tax_expense = CalculatedExpense('tax_expense', 0, assets, infl)

model = Model(1964, 2025, assets, expenses, tax_rate, tax_expense)
yearly_values = model.simulate(95)
sns.lineplot(yearly_values, dashes=False, markers=True)
```
